```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from hmmlearn import hmm, vhmm
import joblib
import seaborn as sns
import networkx as nx
```

# Comparaison entre les modeles

```python
X_3box = pd.pandas.read_csv("../../data/3Boxes_coords.csv",  names=None)
X_3box
```

```python

scores = list()
models = list()
comp = pd.DataFrame(columns=["Nombre d'etat", "Nombre d'iteration", 'Random Index', 'Vraisemblance'])
comp2 = pd.DataFrame(columns=["Nombre d'etat", "Nombre d'iteration", 'Random Index', 'Vraisemblance'])
i = 0

for n_components in range(2, 30):
    for idx in range(50):
        # iter = 5 
        model = hmm.GaussianHMM(n_components=n_components, random_state=idx,covariance_type = "full", n_iter=52)

        model.fit(X_3box)
        models.append(model)
        scores.append(model.score(X_3box))

        new_values = [n_components, 52, 5, model.score(X_3box)]
        comp2.loc[i] = new_values
        i += 1

model = models[np.argmax(scores)]
em_Z = model.predict(X_3box)
print(pd.unique(em_Z))
comp
```

```python

scores = list()
models = list()
comp1 = pd.DataFrame(columns=["Nombre d'etat", "Nombre d'iteration", 'Random Index', 'Vraisemblance'])
i = 0

for n_components in range(2, 20):
    #for n_iter in [5, 10, 20, 30, 40, 50]:
    for idx in range(50):
        # iter = 5
        model = hmm.GaussianHMM(n_components=n_components, random_state=idx, covariance_type = "full", n_iter=52)
        model.fit(X_3box)
        models.append(model)
        scores.append(model.score(X_3box))

        new_values = [n_components, 50, idx, model.score(X_3box)]
        comp1.loc[i] = new_values
        i += 1

model = models[np.argmax(scores)]
em_Z = model.predict(X_3box)
print(pd.unique(em_Z))
comp1
```
