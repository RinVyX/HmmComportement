```python
from hmmlearn import hmm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib
```

# head

```python


X_head = pd.pandas.read_csv("../data/head_coords.csv",  header=None)
model_head = joblib.load('models/EM/hmm_model_head.pkl')

Z_EM = model_head.predict(X_head)
print(type(Z_EM))

samples, _ = model_head.sample(1375)
print(type(samples))

consensus_hidden_states = np.argmax(model_head.predict_proba(samples), axis=1)
print(len(consensus_hidden_states))
print('consensus\n',consensus_hidden_states)
""" print('samples\n',samples)
print('samples\n',len(samples)) """

plt.figure(figsize=(10, 6))
plt.plot(consensus_hidden_states, label='Consensus Hidden States')
plt.xlabel('Time Step')
plt.ylabel('Hidden State')
plt.title('Consensus Hidden States Over Time')
plt.legend()
plt.show()
```
